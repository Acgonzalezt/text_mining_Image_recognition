```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mantener_canales_color(imagen, color):
    color = max(1, min(color, 30))
    mask = np.zeros_like(imagen)
    if color == 1: ### canal azul
        mask[:, :, 0] = imagen[:, :, 0] 
    elif color == 2:### canal verde
        mask[:, :, 1] = imagen[:, :, 1]  
    elif color == 3: ### canal rojo
        mask[:, :, 2] = imagen[:, :, 2]  
    elif color == 10:### canal rojo y verde
        mask[:, :, 0] = imagen[:, :, 0]  
        mask[:, :, 1] = imagen[:, :, 1]  
    elif color == 20: ### Canal verde y azul
        mask[:, :, 1] = imagen[:, :, 1]  
        mask[:, :, 2] = imagen[:, :, 2]  
    elif color == 30: ### Canal rojo y azul
        mask[:, :, 0] = imagen[:, :, 0]  
        mask[:, :, 2] = imagen[:, :, 2]

    return mask
```

```python
img_1=cv2.imread("imagen1.jpg")
img_1=cv2.cvtColor(img_1,cv2.COLOR_BGR2RGB)
plt.imshow(img_1)
plt.show()
```

```python
imagen_con_color = mantener_canales_color(img_1, color=1)
cv2.imwrite("Imagen_canal.jpg", imagen_con_color)
Imagen_canal=cv2.imread("Imagen_canal.jpg")
Imagen_canal=cv2.cvtColor(Imagen_canal,cv2.COLOR_BGR2RGB)
plt.imshow(Imagen_canal)
plt.show()
```

```python
imagen_con_color = mantener_canales_color(img_1, color=10)
cv2.imwrite("Imagen_canal.jpg", imagen_con_color)
Imagen_canal=cv2.imread("Imagen_canal.jpg")
Imagen_canal=cv2.cvtColor(Imagen_canal,cv2.COLOR_BGR2RGB)
plt.imshow(Imagen_canal)
plt.show()
```

```python
imagen_con_color = mantener_canales_color(img_1, color=30)
cv2.imwrite("Imagen_canal.jpg", imagen_con_color)
Imagen_canal=cv2.imread("Imagen_canal.jpg")
Imagen_canal=cv2.cvtColor(Imagen_canal,cv2.COLOR_BGR2RGB)
plt.imshow(Imagen_canal)
plt.show()
```

```python
def reconstruir_imagen_color(escala_gris_rojo, escala_gris_verde, escala_gris_azul):
    if escala_gris_rojo is None or escala_gris_verde is None or escala_gris_azul is None:
        raise ValueError("Al menos una de las imágenes en escala de grises es inválida.")
    if escala_gris_rojo.shape != escala_gris_verde.shape or escala_gris_rojo.shape != escala_gris_azul.shape:
        raise ValueError("Las imágenes en escala de grises deben tener las mismas dimensiones.")
    imagen_color = np.stack((escala_gris_rojo, escala_gris_verde, escala_gris_azul), axis=-1)
    return imagen_color  
```

```python
imagen_gris_rojo = cv2.imread('imagen1_salida_gray_rojo.jpg', cv2.IMREAD_GRAYSCALE)
imagen_gris_verde = cv2.imread('imagen1_salida_gray_verde.jpg', cv2.IMREAD_GRAYSCALE)
imagen_gris_azul = cv2.imread('imagen1_salida_gray_azul.jpg', cv2.IMREAD_GRAYSCALE)

if imagen_gris_rojo is None or imagen_gris_verde is None or imagen_gris_azul is None:
    print("Error al leer al menos una de las imágenes en escala de grises.")
else:
    imagen_color = reconstruir_imagen_color(imagen_gris_rojo, imagen_gris_verde, imagen_gris_azul)
    cv2.imwrite('Imagen a color reconstruida.jpg', imagen_color)
    Imagen_reconstruida=cv2.imread("Imagen a color reconstruida.jpg")
Imagen_reconstruida=cv2.cvtColor(Imagen_reconstruida,cv2.COLOR_BGR2RGB)
plt.imshow(Imagen_reconstruida)
plt.show()   
```

```python
imagen_entrada=cv2.imread("perritos.jpg")
imagen_entrada=cv2.cvtColor(imagen_entrada,cv2.COLOR_BGR2RGB)
plt.imshow(imagen_entrada)
plt.show()
```

```python
def crear_escala_grises_3D(imagen):
    alto, ancho, _ = imagen.shape

    canal_azul = imagen[:, :, 0]
    canal_verde = imagen[:, :, 1]
    canal_rojo = imagen[:, :, 2]

    escala_gris_azul = cv2.cvtColor(canal_azul, cv2.COLOR_GRAY2BGR)
    escala_gris_verde = cv2.cvtColor(canal_verde, cv2.COLOR_GRAY2BGR)
    escala_gris_rojo = cv2.cvtColor(canal_rojo, cv2.COLOR_GRAY2BGR)

    return escala_gris_azul, escala_gris_verde, escala_gris_rojo
```

```python
if imagen_entrada is not None:
    escala_gris_azul, escala_gris_verde, escala_gris_rojo = crear_escala_grises_3D(imagen_entrada)

    cv2.imwrite('escala_gris_azul.jpg', escala_gris_azul)
    cv2.imwrite('escala_gris_verde.jpg', escala_gris_verde)
    cv2.imwrite('escala_gris_rojo.jpg', escala_gris_rojo)
    
else:
    print("Error al leer la imagen original.")
```
